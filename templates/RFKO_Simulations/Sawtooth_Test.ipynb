```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

### Random sawtooth signal

```python
t = np.linspace(1, 3, 2000)
sample = 1e-3*signal.sawtooth(2*np.pi*t, width=1) + 0.666
```

### Plot

```python
plt.plot(t, sample)
```

### Build a Sawtooth signal

```python
t_start = 1.5e-3
t_end = 3e-3
n_points = 2000
n_peaks = 7
duration = t_end - t_start

# Sawtooth period
T = duration / n_peaks
f_saw = 1 / T  # Frequency in Hz

t = np.linspace(t_start, t_end, n_points)

min_freq = 0.665
max_freq = 0.667

# Scale sawtooth to [min_freq, max_freq]
fr_range = max_freq - min_freq
fr_mid = min_freq + fr_range / 2

saw_values = (fr_range / 2) * signal.sawtooth(2 * np.pi * f_saw * t, width=1.0) + fr_mid
```

### Plot

```python
plt.figure(figsize=(10,6))
plt.plot(t, saw_values)
plt.ylim(0.664, 0.668)
plt.xlabel('t (s)')
plt.ylabel('frequency (fractional tune)')
plt.grid()
```

### Build a triangular signal

```python
# Width is set to 0.5
triang_values = (fr_range / 2) * signal.sawtooth(2 * np.pi * f_saw * t, width=0.5) + fr_mid
```

### Plot

```python
plt.figure(figsize=(10,6))
plt.plot(t, triang_values)
plt.ylim(0.664, 0.668)
plt.xlabel('t (s)')
plt.ylabel('frequency (fractional tune)')
plt.grid()
```

### Build an inverted sawtooth signal

```python
# Width is set to 0
triang_values = (fr_range / 2) * signal.sawtooth(2 * np.pi * f_saw * t, width=0) + fr_mid
```

### Plot

```python
plt.figure(figsize=(10,6))
plt.plot(t, triang_values)
plt.ylim(0.664, 0.668)
plt.xlabel('t (s)')
plt.ylabel('frequency (fractional tune)')
plt.grid()
```
