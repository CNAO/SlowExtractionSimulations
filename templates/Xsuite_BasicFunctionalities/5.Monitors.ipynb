```python
import numpy as np
import matplotlib.pyplot as plt
import xtrack as xt
```

```python
line = xt.Line.from_json('../slow_extraction_tutorial/pimms_00_optics.json')
```

```python
num_particles = 50
monitor_ip1 = xt.ParticlesMonitor(start_at_turn=27, stop_at_turn=30,
                                    num_particles=num_particles, auto_to_numpy=True)
line.insert_element(index='drift_2', element=monitor_ip1, name='mymon1')
```

```python
tt = line.get_table()
tt.show()
```

```python
line.build_tracker()
```

```python
x_gen = np.linspace(0, 2.5e-2, 50)
particles = line.build_particles(num_particles=num_particles, x=x_gen,px=0,y=0,py=0,delta=0)

num_turns = 30
line.track(particles, num_turns=num_turns)
```
